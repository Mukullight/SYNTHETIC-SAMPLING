```python
import pymongo
import pandas as pd
df = pd.read_csv("fraud.csv")
class MongoDBHandler:
    def __init__(self, host, port, database_name, collection_name):
        self.client = pymongo.MongoClient(f"mongodb://{host}:{port}")
        self.db = self.client[database_name]
        self.collection = self.db[collection_name]

    def insert_data(self, data):
        if isinstance(data, list):
            self.collection.insert_many(data)
        elif isinstance(data, dict):
            self.collection.insert_one(data)
        else:
            raise ValueError("Data should be a list or a dictionary.")

    def retrieve_data_as_dataframe(self, query=None):
        cursor = self.collection.find(query) if query else self.collection.find()
        data = list(cursor)
        df = pd.DataFrame(data)
        return df

    def close_connection(self):
        self.client.close()

if __name__ == "__main__":
    data = df.head(100).to_dict(orient="records")
    mongo_handler = MongoDBHandler(host="localhost", port=27017, database_name="fraud_detection_system", collection_name="data")
    mongo_handler.insert_data(data)
    mongo_handler = MongoDBHandler(host="localhost", port=27017, database_name="fraud_detection_system", collection_name="data")
    retrieved_data = mongo_handler.retrieve_data_as_dataframe()
```
